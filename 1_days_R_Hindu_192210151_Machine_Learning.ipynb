```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

data = load_iris()
X = data.data
y = data.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)

models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier()
}

for name,model in models.items():
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    acc = accuracy_score(y_test,pred)
    print(name,"Accuracy:",acc)
```

```python
import numpy as np

data = np.array([
['Sunny','Warm','Normal','Strong','Warm','Same','Yes'],
['Sunny','Warm','High','Strong','Warm','Same','Yes'],
['Rainy','Cold','High','Strong','Warm','Change','No'],
['Sunny','Warm','High','Strong','Cool','Change','Yes']
])

X = data[:,:-1]
y = data[:,-1]

hypothesis = ['0'] * len(X[0])

for i in range(len(X)):

    if y[i] == "Yes":
        for j in range(len(hypothesis)):

            if hypothesis[j] == '0':
                hypothesis[j] = X[i][j]

            elif hypothesis[j] != X[i][j]:
                hypothesis[j] = '?'

print("Most Specific Hypothesis:")
print(hypothesis)
```

```python
import pandas as pd

data = pd.DataFrame({
'Sky':['Sunny','Sunny','Rainy','Sunny'],
'Temp':['Warm','Warm','Cold','Warm'],
'Humidity':['Normal','High','High','High'],
'Wind':['Strong','Strong','Strong','Strong'],
'Water':['Warm','Warm','Warm','Cool'],
'Forecast':['Same','Same','Change','Change'],
'EnjoySport':['Yes','Yes','No','Yes']
})

X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

S = ['0'] * len(X[0])
G = ['?'] * len(X[0])

for i in range(len(X)):

    if y[i] == 'Yes':
        for j in range(len(S)):
            if S[j] == '0':
                S[j] = X[i][j]
            elif S[j] != X[i][j]:
                S[j] = '?'

    elif y[i] == 'No':
        for j in range(len(S)):
            if S[j] != X[i][j]:
                G[j] = S[j]
            else:
                G[j] = '?'

print("Specific Hypothesis S:", S)
print("General Hypothesis G:", G)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

data = pd.DataFrame({
'Outlook':['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny'],
'Temperature':['Hot','Hot','Hot','Mild','Cool','Cool','Mild','Cool'],
'Humidity':['High','High','High','High','Normal','Normal','Normal','High'],
'Wind':['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak'],
'PlayTennis':['No','No','Yes','Yes','Yes','No','Yes','No']
})

le = LabelEncoder()

for column in data.columns:
    data[column] = le.fit_transform(data[column])

X = data.iloc[:,:-1]
y = data.iloc[:,-1]

model = DecisionTreeClassifier(criterion="entropy")
model.fit(X,y)

sample = [[2,1,1,0]]

prediction = model.predict(sample)

print("Prediction for new sample:", prediction)
```
